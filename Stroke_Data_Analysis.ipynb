# Stroke Dataset

```python
# import our libraries
import pandas as pd
```

```python
df = pd.read_csv("stroke_data.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(columns=['id'],inplace = True)
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['bmi'].fillna(df['bmi'].median(), inplace=True)
df
```

```python
missing_percent = df.isnull().mean() * 100
print(missing_percent)
```

```python
print(df['smoking_status'].value_counts())
```

```python
df = df[df['smoking_status'] != 'unknown']
```

```python
bins = [0, 18, 35, 50, 65, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior', 'Elderly']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df
```

```python
bmi_bins = [0, 18.5, 24.9, 29.9, 100]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['bmi_category'] = pd.cut(df['bmi'], bins=bmi_bins, labels=bmi_labels)
```

```python
df
```

```python
mean_age = df['age'].mean()
median_bmi = df['bmi'].median()
print(f"Mean Age: {mean_age:.2f}")
print(f"Median BMI: {median_bmi:.2f}")
```

```python
min_glucose = df['avg_glucose_level'].min()
max_glucose = df['avg_glucose_level'].max()
print(f"Glucose level ranges from {min_glucose} to {max_glucose}")
```

```python
std_age = df['age'].std()
var_bmi = df['bmi'].var()
print(f"Standard deviation of age: {std_age:.2f}")
print(f"Variance of BMI: {var_bmi:.2f}")
```

```python
stroke_counts = df.groupby('work_type')['stroke'].count()
print(stroke_counts)
```

```python
agg_stats = df.groupby('stroke').agg({
    'age': ['mean', 'median', 'min', 'max'],
    'avg_glucose_level': ['mean', 'std']
})
print(agg_stats)
```

```python
print(df['smoking_status'].value_counts())
print(df['gender'].value_counts())
print(df['age_group'].value_counts())
```

```python
df
```

# Shows the distribution of stroke cases across genders.

```python
import matplotlib.pyplot as plt

gender_stroke = df.groupby(['gender', 'stroke']).size().unstack(fill_value=0)

gender_stroke.plot(kind='bar', stacked=False)
plt.title('Stroke Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['No Stroke', 'Stroke'])
plt.tight_layout()
plt.show()
```

# Compare age distributions between stroke and non-stroke groups.


```python
plt.figure(figsize=(10,6))

plt.hist(df[df['stroke'] == 0]['age'], bins=30, alpha=0.6, label='No Stroke', color='skyblue')
plt.hist(df[df['stroke'] == 1]['age'], bins=30, alpha=0.6, label='Stroke', color='salmon')

plt.title('Age Distribution by Stroke Outcome')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Visualizes glucose levels’ variation between stroke and non-stroke patients.


```python
glucose_stroke_0 = df[df['stroke'] == 0]['avg_glucose_level']
glucose_stroke_1 = df[df['stroke'] == 1]['avg_glucose_level']

plt.figure(figsize=(8,6))
plt.boxplot([glucose_stroke_0, glucose_stroke_1], labels=['No Stroke', 'Stroke'])
plt.title('Average Glucose Level by Stroke Outcome')
plt.ylabel('Average Glucose Level')
plt.show()
```

# Shows stroke incidence across smoking categories.

```python
stroke_smoking = df.groupby('smoking_status')['stroke'].mean()

plt.figure(figsize=(8,6))
plt.bar(stroke_smoking.index, stroke_smoking.values, color='mediumseagreen')
plt.title('Stroke Rate by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Stroke Rate')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

# Visualizes the proportion of patients in each BMI category.

```python
bmi_counts = df['bmi_category'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'gold', 'lightgreen'])
plt.title('BMI Category Distribution')
plt.axis('equal')  # Equal aspect ratio for perfect circle
plt.show()
```

# Shows how stroke rates differ across age groups.

```python
stroke_age_group = df.groupby('age_group')['stroke'].mean()

plt.figure(figsize=(8,6))
plt.bar(stroke_age_group.index, stroke_age_group.values, color='orange')
plt.title('Stroke Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Stroke Rate')
plt.tight_layout()
plt.show()
```

# Shows stroke counts or rates for different work types.

```python
work_stroke = df.groupby('work_type')['stroke'].mean().sort_values()

plt.figure(figsize=(10,6))
plt.barh(work_stroke.index, work_stroke.values, color='purple')
plt.title('Stroke Rate by Work Type')
plt.xlabel('Stroke Rate')
plt.ylabel('Work Type')
plt.tight_layout()
plt.show()
```

# Insight: Shows how stroke risk varies by age and gender — often stroke risk increases with age, and gender differences may appear.

```python
import numpy as np

age_gender_stroke = df.groupby(['age_group', 'gender'])['stroke'].mean().unstack()
age_groups = age_gender_stroke.index
male_rates = age_gender_stroke['Male']
female_rates = age_gender_stroke['Female']

x = np.arange(len(age_groups))
width = 0.35

plt.figure(figsize=(10,6))
plt.bar(x - width/2, male_rates, width, label='Male', color='steelblue')
plt.bar(x + width/2, female_rates, width, label='Female', color='lightcoral')

plt.xticks(x, age_groups)
plt.ylabel('Stroke Rate')
plt.title('Stroke Rate by Age Group and Gender')
plt.legend()
plt.tight_layout()
plt.show()
```

# Insight: How combined presence of hypertension and heart disease affects stroke rate.

```python
risk_group = df.groupby(['hypertension', 'heart_disease'])['stroke'].mean().unstack()

risk_group.plot(kind='bar', figsize=(8,6))
plt.title('Stroke Rate by Hypertension and Heart Disease')
plt.xlabel('Hypertension (0=No,1=Yes)')
plt.ylabel('Stroke Rate')
plt.legend(title='Heart Disease (0=No,1=Yes)')
plt.tight_layout()
plt.show()
```

# Insight: Explore interaction effect of occupation and smoking on stroke risk.

```python
work_smoking_stroke = df.groupby(['work_type', 'smoking_status'])['stroke'].mean().unstack()

work_smoking_stroke.plot(kind='bar', figsize=(12,7))
plt.title('Stroke Rate by Work Type and Smoking Status')
plt.xlabel('Work Type')
plt.ylabel('Stroke Rate')
plt.xticks(rotation=45)
plt.legend(title='Smoking Status')
plt.tight_layout()
plt.show()
```

# Insight: Explore how living environment and marital status relate to stroke risk.

```python
residence_married_stroke = df.groupby(['Residence_type', 'ever_married'])['stroke'].mean().unstack()

residence_married_stroke.plot(kind='bar', figsize=(8,6))
plt.title('Stroke Rate by Residence Type and Marital Status')
plt.xlabel('Residence Type')
plt.ylabel('Stroke Rate')
plt.legend(title='Ever Married')
plt.tight_layout()
plt.show()
```

# Insight: Visualize relationship between age, glucose, and stroke occurrence.

```python
colors = df['stroke'].map({0: 'blue', 1: 'red'})

plt.figure(figsize=(10,6))
plt.scatter(df['age'], df['avg_glucose_level'], c=colors, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.title('Age vs Average Glucose Level Colored by Stroke')
plt.legend(['No Stroke (blue)', 'Stroke (red)'])
plt.show()
```

#  Distribution of BMI Categories by Stroke Status (Stacked Bar Chart)

```python
bmi_stroke_counts = pd.crosstab(df['bmi_category'], df['stroke'])

bmi_stroke_counts.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title('Distribution of BMI Categories by Stroke Status')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.legend(['No Stroke', 'Stroke'])
plt.tight_layout()
plt.show()
```

# Calculate stroke rate cumulatively by sorting age ascending — shows risk increase over lifespan.

```python
df_sorted = df.sort_values('age')
df_sorted['cumulative_stroke_rate'] = df_sorted['stroke'].expanding().mean()

plt.figure(figsize=(10,6))
plt.plot(df_sorted['age'], df_sorted['cumulative_stroke_rate'], color='darkgreen')
plt.title('Cumulative Stroke Rate by Increasing Age')
plt.xlabel('Age')
plt.ylabel('Cumulative Stroke Rate')
plt.grid(True)
plt.show()
```

# Shows age spread per work type, which can correlate with stroke risk.

```python
work_types = df['work_type'].unique()
data = [df[df['work_type'] == work]['age'] for work in work_types]

plt.figure(figsize=(12,6))
plt.boxplot(data, labels=work_types, patch_artist=True)
plt.title('Age Distribution by Work Type')
plt.ylabel('Age')
plt.xlabel('Work Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Stroke Rate by Gender and Smoking Status (Grouped Bar Chart)

```python
gender_smoking_stroke = df.groupby(['gender', 'smoking_status'])['stroke'].mean().unstack()

gender_smoking_stroke.plot(kind='bar', figsize=(10,6))
plt.title('Stroke Rate by Gender and Smoking Status')
plt.xlabel('Gender')
plt.ylabel('Stroke Rate')
plt.xticks(rotation=0)
plt.legend(title='Smoking Status')
plt.tight_layout()
plt.show()
```

# Boxplot of Average Glucose Level by Work Type

```python
plt.figure(figsize=(12,6))
glucose_data = [df[df['work_type'] == w]['avg_glucose_level'] for w in work_types]

plt.boxplot(glucose_data, labels=work_types, patch_artist=True)
plt.title('Average Glucose Level by Work Type')
plt.ylabel('Average Glucose Level')
plt.xlabel('Work Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
